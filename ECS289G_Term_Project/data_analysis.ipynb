```python
# ECS289G_Term_Project/data_analysis.ipynb

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"/Users/harshil/Developer/GitHub_Repos/ECS_289G/data/processsed/cleaned_dataset.csv"
df = pd.read_csv(file_path, low_memory=False)
```

```python
# Display dataset information
print(f"Dataset Shape: {df.shape}")
df.head()
df.info()
```

```python
# Display columns in the dataset
print("Dataset Columns:")
print(df.columns)
```

```python
# Missing values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
missing_summary = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage': missing_percentage})
missing_summary = missing_summary[missing_summary['Missing Count'] > 0]
print("Missing Values Summary:")
print(missing_summary)
```

```python
# Definining feature groups
numerical_features = [
    'tract_to_msa_income_percentage', 
    'ffiec_msa_md_median_family_income', 
    'tract_minority_population_percent', 
    'interest_rate'
]

categorical_features = [
    'loan_type_2', 'loan_type_3', 'loan_type_4',
    'loan_purpose_2', 'loan_purpose_4', 'loan_purpose_5', 
    'loan_purpose_31', 'loan_purpose_32', 
    'lien_status_2', 'construction_method_2', 
    'occupancy_type_2', 'occupancy_type_3'
]

demographic_features = [
    'race_0', 'race_1', 'race_2', 'race_3', 'race_4', 
    'race_5', 'race_6', 'race_7', 'race_8', 
    'gender_0', 'gender_1', 'gender_2', 'gender_3', 
    'ethnicity_0', 'ethnicity_1', 'ethnicity_2', 
    'ethnicity_3', 'ethnicity_4'
]

# Seting target variable
target_variable = 'action_taken'
```

```python
# Distribution of numerical features
for col in numerical_features:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True, bins=30, color='blue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[numerical_features].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python
# Distribution of categorical features
for col in categorical_features:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=df[col], order=df[col].value_counts().index, palette="viridis")
    plt.title(f"Value Counts of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Distribution of demographic features
for col in demographic_features:
    plt.figure(figsize=(10, 5))
    sns.barplot(x=df[col].value_counts().index, y=df[col].value_counts(), palette="cool")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Target variable distribution
plt.figure(figsize=(10, 5))
sns.countplot(x=df[target_variable], order=df[target_variable].value_counts().index, palette="Blues")
plt.title(f"Distribution of {target_variable}")
plt.xlabel("Action Taken")
plt.ylabel("Frequency")
plt.show()
```
